Scraping From SIT Website

```python
from bs4 import BeautifulSoup

import requests
url = "https://www.sitpune.edu.in/"
```

```python
response = requests.get(url)
print (response)
```

```python
html_content = response.content
```

```python
soup = BeautifulSoup(html_content, "html.parser")
```

Returns the entire title tag

```python
soup.find('title')
```

Returns the string of the title tag

```python
soup.find('title').string
```

```python
h1_tags = soup.find_all('h1')
h1_tags
```

```python
for h1 in h1_tags:
    print(h1.text)
```

Exrtracting <a> tags

```python
a_tags= soup.find_all('a')
a_tags
```

Gives the number of a tags on the page

```python
len(a_tags)
```

```python
a_tags
```

Extracting links from a tags

```python
list_of_links = []
for a in a_tags:
    link = a.get('href')
    if link != None:
      list_of_links.append(link)

```

Gives the number of links passed in the a tags on the page

```python
len(list_of_links)
```

Displays the list of links

```python
list_of_links
```

Creates a dataframe with number of links and columns

```python
df_links= pd.DataFrame(data=list_of_links, columns=['Links'])
```

```python
img_links= []
link = soup.select('img[src="https://www.sitpune.edu.in/assets/image"]')
```

```python
for img in link:
    img_links.append(img.get('src'))
```

```python
directory = '/content/Scrap images'
```

```python
for index, img_link in enumerate(img_links):
    img_data = requests.get(img_link).content
    with open(f'{directory}/'+str(index+1)+'.jpg', 'wb+') as f:
        f.write(img_data)
```
